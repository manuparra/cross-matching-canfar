# Crossmatching

## Setup

First, import the crossmatch module.

```python
import sys
sys.path.append('/opt/cross-matching/scripts/')
from crossmatch import Crossmatch
```

Then, set some parameters for the crossmatching:

```python
crossmatch_parameters = {
    'ra': 192.25,
    'dec': 25,
    'radius': 1,
    'ref_catalogue': 'GAIA_DR3',
    'match_catalogues': ['LOTSS_DR2', 'TGSS_ADR', '2MASS'],
    'max_separation': 0.02
}
```

And instantiate a Crossmatch with these parameters.

```python
crossmatch = Crossmatch(**crossmatch_parameters)
```

## Running

```python
crossmatch_results = crossmatch.run()
crossmatch_results
```

### Plotting

```python
crossmatch.plot(crossmatch_results=crossmatch_results)
```
